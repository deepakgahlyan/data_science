```python
from bs4 import BeautifulSoup as bs
```

```python
import requests as rq
data= rq.get('https://en.wikipedia.org/wiki/List_of_countries_by_coffee_production')
html_text=data.text
html_text
```

```python
soup=bs(html_text)
```

```python
soup
```

```python
table_val=soup.find('table')
table_val
```

```python
for i in table_val:
    print(i.text)
```

```python
# as output is not as desired table so wee need to arrange the data as table 
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(4,4))
lst=[]
for i in table_val.find_all('tr'):
    r_data=i.text
    data=r_data.replace('\n\n','-').strip()
    lst.append(data.split('-'))
df=pd.DataFrame(lst[1:],columns=lst[0])
df
```

```python
datatype=df['Pounds'].values
print(datatype)
print(type(datatype[0])) # this show class of the value data type which is a string so now i need to convert this string into 
# int 
```

```python
# clearly  , is not numerical so can't to type casting directly 
lst=[]
for i in datatype:
    j=i.split(',')  # to split via , and make a list 
    k=''.join(j)   # to convert in string as it will not have ,
    lst.append(int(k))
pounds=pd.Series(lst)
print(pounds)
```

```python
# we will do same for '60 kilogram' and Metric tons
datatype=df['60 kilogram bags'].values
lst=[]
for i in datatype:
    j=i.split(',')  # to split via , and make a list 
    k=''.join(j)   # to convert in string as it will not have ,
    lst.append(int(k))
kg_60=pd.Series(lst)
print(kg_60)
```

```python
datatype=df['Metric tons'].values
lst=[]
for i in datatype:
    j=i.split(',')  # to split via , and make a list 
    k=''.join(j)   # to convert in string as it will not have ,
    lst.append(int(k))
metric_tons=pd.Series(lst)
print(metric_tons)
```

```python
df['60 kilogram bags']=kg_60
df['Metric tons']=metric_tons
df['Pounds']=pounds
```
